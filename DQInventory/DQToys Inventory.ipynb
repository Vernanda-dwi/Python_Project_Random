```python
# Membaca datad engan memasukkannya ke variabel data

import pandas as pd

data = pd.read_excel("DQToys Inventory.xlsx")

data
```

```python
# info keseluruhan kolom
data.info()
```

```python
# Mnecari missing value pada data ini
msv_data = data.isnull().sum()
msv_data
```

Tidak ditemukan duplikat pada data

```python
dpl1 = data.duplicated().sum()
print(f"Jumlah data duplikat: {dpl1}")
```

```python
#mencari data duplikat
dpl_data = data.duplicated(keep=False).sum()
print(f"Jumlah data duplikat: {dpl_data}")
```

```python
# mengapus baris di kolom data id yang memeiliki missing value

data2 = data.dropna(subset=['sale_id'])

msv_data2 = data2.isnull().sum()
msv_data2
```

```python
# download data yang sudah di cleanung
data2.to_csv('data_cleaned.csv', index=False)
```
